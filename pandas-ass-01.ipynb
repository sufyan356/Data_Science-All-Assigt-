```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\Windows10\Desktop\DATA_Science_SMIT\Data\covid_19_country_wise_latest.csv')
data.head()
```

```python
df = data.copy()
```

```python
#Q:01 Load and Display Data → Read the dataset and print the first 5 rows.
df.head()
```

```python
# Q:02: Set the Dataset index is Unnamed: 0
df.set_index("Unnamed: 0",inplace=True)
```

```python
df.head()
```

```python
# Q:03: Now replace the index column Unnamed: 0 name with index
df.index.name = 'index'
```

```python
df.head()
```

```python
# Q:04: Create Function that take dataset columns and replace whitespace 
# with Underscore and update dataset automatically and the pass the 
# dataset.

df.columns=df.columns.str.replace(" ", "_")
df.columns
```

```python
#Q:05 Check Basic Information → Find column names, data types, and 
# missing values.

df.info()
```

```python
#Q:06 fill the nan values with the mean of the column

df.fillna({
    'New_cases': df['New_cases'].mean(),
    'New_deaths': df['New_deaths'].mean(),
},inplace=True)

df.isnull().sum()
```

```python
#Q:07: Count the number of unique countries in the dataset.
for x,y in df['Country/Region'].value_counts().items():
    print(x,y)
```

```python
#Q:08: Check if there are any duplicate country entries and remove them if 
# needed.

df[df['Country/Region'].duplicated()]
```

```python
#Q:09 Find the mean, median, and standard deviation of total cases
mean = df['Confirmed'].mean()
print(f'mean of total cases: {mean}')
```

```python
median = df['Confirmed'].median()
print(f'median of total cases: {median}')
```

```python
std = df['Confirmed'].std()
print(f'std of total cases: {std}')
```

```python
# Q:10: Find out the string in Deaths Column and replace it with the 
# mean of Deaths column.

# Convert column to number, replace alphabetic values with NaN
df['Deaths'] = pd.to_numeric(df['Deaths'],errors='coerce')
```

```python
death_mean = df['Deaths'].mean()
```

```python
df['Deaths'].fillna(death_mean,inplace=True)
```

```python
df['Deaths'].unique()
```

```python
#Q:11 Change the datatype of column Deaths
df['Deaths'].astype(int)
```

```python
# Q:12. Calculate total number of Death and Recovery all over the 
# world
```

```python
df[['Deaths','Recovered']].sum()
```

```python
#Q: 13. How many countries have more than 1 million total cases
df[df['Confirmed'] > 1000000][['Confirmed','Country/Region']]
```

```python
#Q: 14. Which countries have a recovery rate (Recovered / Total Cases) 
# above 95%
df['recovery_rate'] = round((df['Recovered']/df['Confirmed'])*100,2)
df[df['recovery_rate']>95][['Country/Region','recovery_rate']].sort_values(by='recovery_rate', ascending=False)
```

```python
# Q: 15. Drop columns name WHO Region and Confirmed

df.drop(columns=['Confirmed','WHO_Region'],inplace=True)
df.head(2)
```

```python
# Q: 16. Find the Country with the Max Deaths
df[df['Deaths']==df['Deaths'].max()]['Country/Region'].values[0]
```

```python
# Q: 17. Sort Countries by Deaths (Descending Order)
df[['Country/Region','Deaths']].sort_values(by='Deaths',ascending=False)
```

```python
# Q:18. Make new column name Total_cases that have a sum of Deaths, 
# Recovered , Active

df['Total_cases'] = df['Deaths'] + df['Recovered'] + df['Active']
```

```python
# Q:19. Calculate Death_Rate for Each Country (Deaths per Total 
# Cases) by suing formula (Desths/Total_cases)*100 and save it in 
# Dath_Rate column

df['Dath_Rate']=(df['Deaths']/df['Total_cases'])*100
```

```python
# Q:20. Identify countries where total cases are increasing but death 
# rates remain low
# df[['Country/Region','Total_cases','Dath_Rate']].sort_values(by='Total_cases' , ascending=False)
low_death_rate_thresh = 2
for i in df[df['Total_cases'] > low_death_rate_thresh]['Country/Region'].values:
    print(i)
```

```python
# Q: 21. print head of only 2 columns Country/Region and Death_Rate

df[['Country/Region','Dath_Rate']].head(2)
```

```python
# Q: 22. Create a scatter plot comparing total cases and total deaths.
plt.figure(figsize=(10,4))
sns.scatterplot(x=df['Total_cases'], y=df['Deaths'])
plt.title('Comparing total cases and total deaths')
plt.show()
```

```python
#Q:23. Save dataset in CSV format.
df.to_csv(r'C:\Users\Windows10\Desktop\DATA_Science_SMIT\Data\covid19_cleaned.csv')

#---------------------------------- end ---------------------------------------
```
